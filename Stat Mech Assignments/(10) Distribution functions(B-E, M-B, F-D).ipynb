```python
import numpy as np  
import matplotlib.pyplot as plt 


e = 1.6e-19  
k = 1.38e-23

E = np.linspace(-0.5,0.5,1000)
for T in range(100,1300,200):
    
    def f(a,T):  
        return 1/((np.exp(((E)*e)/(k*T)))+a)  

    plt.subplot(2,3,1)  
    plt.plot(E,f(-1,T),label='T = {} K'.format(T))  
    plt.xlim(0,1)  
    plt.ylim(0,2)  
    plt.xlabel('Energy(eV)')  
    plt.ylabel('f(E)')  
    plt.legend(loc='best',prop={'size':6})  
    plt.title("B-E distribution") 
    
    plt.subplot(2,3,3)  
    plt.plot(E,f(1,T),label='T = {} K'.format(T))  
    plt.legend(loc='best',prop={'size':5})  
    plt.ylim(0,1)  
    plt.xlabel('Energy(eV)')  
    plt.ylabel('f(E)')  
    plt.title("F-D distribution")  
    
    plt.subplot(2,3,5)  
    plt.plot(E,f(0,T),label='T = {} K'.format(T))   
    plt.ylim(0,3) 
    plt.xlabel('Energy(eV)')  
    plt.ylabel('f(E)')  
    plt.legend(loc='best',prop={'size':5})  
    plt.title("M-B distribution") 
    
plt.show()     
```
