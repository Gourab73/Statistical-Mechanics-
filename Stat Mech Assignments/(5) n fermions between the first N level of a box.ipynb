### Canonical partition function of n fermions distributed between the first N levels of a box :

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt

def partition_function(n, N, y, beta):
    E_level = [y / (k**2) for k in range(1, N + 1)]
    combinations = list(itertools.combinations(E_level, n))
    partition_function = sum([np.prod(np.exp(-beta * sum(combination))) for combination in combinations])
    return partition_function
n = 2
N = 4
y = 1.0
beta = 1.0

beta_values = np.linspace(0.1, 10.0, 100)

partition_function_values = [partition_function(n, N, y, beta) for beta in beta_values]

plt.plot(beta_values, partition_function_values)
plt.xlabel(r'$\beta$')
plt.ylabel('Partition Function')
plt.grid()
plt.title('Canonical Partition Function')
plt.show()
```
