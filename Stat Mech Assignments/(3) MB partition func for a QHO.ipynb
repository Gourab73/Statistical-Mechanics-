### M-B partition function for a Quantum Harmonic Oscillator :

```python
import numpy as np
import matplotlib.pyplot as plt

def harmonic_oscillator(beta, E_levels):
    partition_function = np.sum(np.exp(-beta * E_levels))
    return partition_function

beta = 1.0
n_max = 105 
hbar_w = 1

E_levels = (np.arange(n_max) + 0.5) * hbar_w

partition_function = harmonic_oscillator(beta, E_levels)

x = np.arange(n_max)
y = np.exp(-beta * E_levels) / partition_function

plt.figure(figsize=(5, 4))
plt.plot(x, y)
plt.xlabel(r"$n$")
plt.ylabel("Probability")
plt.title(r"$E_n = \left(n + \frac{1}{2}\right) \hbar \omega$")
plt.grid()
plt.show()
```
